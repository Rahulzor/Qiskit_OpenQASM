```python
!pip install qiskit qiskit-aer -q


from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import qiskit.qasm3 as qasm3

qc = QuantumCircuit(3)
qc.h(0)
qc.x(1)
qc.cx(0, 2)
qc.cz(1, 2)
qc.barrier()
qc.measure_all()
print("Modified Quantum Circuit (3 Qubits):")
print(qc.draw())

qasm_code = qasm3.dumps(qc)
print("\nGenerated OpenQASM 3 Code:\n")
print(qasm_code)
print("\nğŸ§  Representation in QASM:")
print("Hadamard gate (H) â†’ represented as:  h q[0];")
print("CNOT gate (CX) â†’ represented as:     cx q[control], q[target];")
print("\nğŸ” Barrier Representation:")
print("In QASM, barrier appears as:")
print("barrier q[0], q[1], q[2];")
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit).result()
counts = result.get_counts()

print("\nSimulation Results:")
print(counts)
```
